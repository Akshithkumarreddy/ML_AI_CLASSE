Reshaping data using pandas

```python
#reshaping data

a = pd.melt(drinks) #melt converts all colums into rows

print(a)
```

```python
a = drinks.pivot(columns = 'country',values ='continent') #pivot converts all rows into columns ; we can provide index also

print(a)
```

```python
a = drinks.country
#print(a)

b = drinks.continent
#print(b)

print(pd.concat([a,b],axis=1).head()) #append colums
```

```python
a = drinks.country[:8]
#print(a)

b = drinks.country[8:]
#print(b)

#print(pd.concat([a,b]).head())
print(pd.concat([a,b],axis=0).head()) #append rows
```
